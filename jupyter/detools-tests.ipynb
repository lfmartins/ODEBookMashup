```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from detools import *
```

```python
# Single equation
def f(t, x):
    return x * (1 - x)
tvec, xvec = solve_de(f, 0.0, 0.5, -0.25, 5.0)
plt.plot(tvec,xvec)
None
```

```python
# Single equation, with arguments
def f(t, x, r, C):
    return r * x * (1 - x / C)
tvec, vec = solve_de(f, 0.0, 0.5, -0.25, 5.0, args=(1, 2))
plt.plot(tvec,xvec)
None
```

```python
# System, no parameters
def f(t, x):
    R, F = x
    a, b, c, d = 0.75, .2, 1.15, .1
    return np.array([a*R - b*R*F, -c*F + d*R*F])
tvec, xvecs = solve_de_system(f, 0.0, [3.0, 4.0], -5.0, 5.0)
```

```python
plt.plot(xvecs[0],xvecs[1])
None
```

```python
plt.plot(tvec, xvecs[0])
plt.plot(tvec, xvecs[1])
None
```

```python
# System, with parameters
def f(t, x, a, b, c, d):
    R, F = x
    return np.array([a*R - b*R*F, -c*F + d*R*F])
tvec, xvecs = solve_de_system(f, 0.0, [3.0, 4.0], -5.0, 5.0, args=(0.75, .2, 1.15, .1))
```

```python
plt.plot(xvecs[0],xvecs[1])
None
```

```python
plt.plot(tvec, xvecs[0])
plt.plot(tvec, xvecs[1])
None
```

```python
def fsystem(t, xyvec, a, b, c, d):
    x, y = xyvec
    return np.array([a * x + b * y, c * x + d * y])
```

```python
plt.figure(figsize=(8,8))
direction_field(fsystem, (-10,10,21), (-10,10,21), args=(0, 1, -1, 0), color='blue')
None
```
