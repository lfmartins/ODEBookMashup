# Exploration of Linear Second Order Equations

In this lab, you will experiment with second order, constant coefficients homogeneous equations:

$$
ay''+by'+cy=0
$$

## 1. Example

We start with an example, which you should use as a model to solve the exercises that are proposed later. Let's consider the equation:

$$
y''+y'-6y=0
$$

### Step 1: Solve the Differential Equation

Solve the solution using pencil and paper (in this lab we do not need to use the computer to solve the differential equation). The characteristic equation is:

$$
\lambda^2+\lambda-6=0
$$

This factors as:

$$
(\lambda-2)(\lambda +3)=0
$$

The solutions are $\lambda_1=2$ and $\lambda_2=-3$, so that we are in _Case 1: real distinct roots_, and the general solution is:

$$
y(x)=c_1e^{2x}+c_2e^{-3x}
$$

### Step 2: Plot Three Solutions

Let's graph three solutions, by choosig different values of $c_1$ and $c_2$. The code to graph the solutions is in the cells below:

```python
c1, c2 = 1, 0
xvalues = np.linspace(-2,2,500)
yvalues = c1*np.exp(2*xvalues) + c2*np.exp(-3*xvalues)
plt.plot(xvalues, yvalues, lw=2, color='green')
plt.title("$y''+y'-6y=0, c_1={:3.1f}, c_2={:3.1f}$".format(c1,c2), 
          fontsize=18)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$y(x)$', fontsize=18)
None
```

```python
c1, c2 = 0, 1
xvalues = np.linspace(-2,2,500)
yvalues = c1*np.exp(2*xvalues) + c2*np.exp(-3*xvalues)
plt.plot(xvalues, yvalues, lw=2, color='green')
plt.title("$y''+y'-6y=0, c_1={:3.1f}, c_2={:3.1f}$".format(c1,c2), 
          fontsize=18)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$y(x)$', fontsize=18)
None
```

```python
c1, c2 = 2, 3
xvalues = np.linspace(-2,3.5,500)
yvalues = c1*np.exp(2*xvalues) + c2*np.exp(-3*xvalues)
plt.plot(xvalues, yvalues, lw=2, color='green')
plt.title("$y''+y'-6y=0, c_1={:3.1f}, c_2={:3.1f}$".format(c1,c2), 
          fontsize=18)
plt.xlabel('$x$', fontsize=18)
plt.ylabel('$y(x)$', fontsize=18)
None
```

### Step 3: Phase Plane Plot

The phase plane plot is a graph of the curve $(y(x),y'(x))$ on the plane. We can generate a phase plane plot containing the three solutions computed above as shown in the cell below. It is necessary to find the formula for $y'(x)$ by directly computing the derivative:

$$
y(x)=c_1e^{2x}+c_2e^{-3x}
$$

$$
y'(x)=2c_1e^{2x}-3c_2e^{-3x}
$$

Here is how to generate the plot. Notice that we plot the three curves in the same set of axes.

```python
cvalues = [[1,0],   [0,1],  [2,3]]
bounds  = [[-2,2],  [-2,2], [-2,3.5]]
colors =  ['green', 'blue', 'red']
for citem, bitem, color in zip(cvalues, bounds, colors):
    c1, c2 = citem
    xmin, xmax = bitem
    xvalues = np.linspace(xmin,xmax, 500)   
    yvalues = c1*np.exp(2*xvalues) + c2*np.exp(-3*xvalues)
    dyvalues = 2*c1*np.exp(2*xvalues) -3*c2*np.exp(-3*xvalues)
    plt.plot(yvalues, dyvalues, lw=2, color=color)
plt.axvline(0, lw=1, color='black')
plt.axhline(0, lw=1, color='black')
plt.title("$y''-y'+6y=0$", fontsize=18)
plt.xlabel('$y$', fontsize=18)
plt.ylabel("$y'$", fontsize=18)
plt.axis([-20, 20, -20, 20])
None
```

### Step 4: Conclusions

We now observe what can be concluded from the plotted solutions:

- $c_1=1, c_2=0$: 
    
    - $\lim_{x\to\infty} y(x)=\infty$
    - $\lim_{x\to-\infty}y(x)=0$
    - Oscillatory: No.
    
- $c_1=0, c_2=1$:

    - $\lim_{x\to\infty} y(x)=0$
    - $\lim_{x\to-\infty}y(x)=\infty$
    - Oscillatory: No.

- $c_1=2, c_2=3$:

    - $\lim_{x\to\infty} y(x)=\infty$
    - $\lim_{x\to-\infty}y(x)=\infty$
    - Oscillatory: No.

## 3. Hints

The solution will usually have an exponential component, which may make it hard to visualize all solutions in the same scale. Use the following strategies to handle this problem.

#### Changing the formulas

You will have to change the formulas to plot the solution $y(x)$, which is stored in the array `yvalues`:

    yvalues = c1*np.exp(2*xvalues) + c2*np.exp(-3*xvalues)
    
In the phase plots, you need to change the formualas for both $y(x)$ and $y'(x)$:

    yvalues = c1*np.exp(2*xvalues) + c2*np.exp(-3*xvalues)
    dyvalues = 2*c1*np.exp(2*xvalues) -3*c2*np.exp(-3*xvalues)

#### Solution plots:

The line of code:

    xvalues = np.linspace(-2,2,500)

sets the $x$ values for which the solution is plotted. In the example above, the solution will be plotted from $x=-2$ to $x=2$. You will need to choose the appropriate bounds in each case.

#### Phase plot:

To specify the solutions and the corresponding bounds, you will have to modify the lines:

    cvalues = [[1,0],   [0,1],  [2,3]]
    bounds  = [[-2,2],  [-2,2], [-2,3.5]]

The first line specifies the values of $c_1$ and $c_2$ for each of the solutions, and the second line the corresponding bounds for $x$. Notice that the bounds may be different for different solutions.

To address the problem of solutions with different scales, modify the line:

    plt.axis([-20, 20, -20, 20])
    
This sets the window to be plotted, in the format `[xmin, xmax, ymin, ymax]`
