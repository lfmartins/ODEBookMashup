# Example 1

Let's consider the family of one-dimensional flows:
$$
\newcommand{\mtab}{\quad\quad\quad\quad}
\mtab\dot{x} = r - x^2
$$
This system has a saddle-node bifurcation at the parameter value $r_c=0$. If $r>0$, it has two equilibrium points, $x^*_1=-\sqrt{r}$ (unstable) and $x^*_2=\sqrt{r}$ (stable).

The curve defining the equilibrium points is, of course,
$$
\mtab r-x^2=0
$$
or
$$
\mtab r=x^2.
$$
Plotting this curve in the $r\times x$ plane is easy enough. We want to be very lazy, so we don't want even to take a square root.

```python
xpos = linspace(0,5,200)
rpos = xpos ** 2
xneg = linspace(-5,0,200)
rneg = xneg ** 2
axis([-25,25,-6,6])
plot(rpos, xpos, color = 'red', linewidth = 2)
plot(rneg, xneg, '--', color = 'red', linewidth = 2)
arrow(-10,1,0,-2, head_width = .4, head_length = .3, color = 'blue')
arrow(7,5.5,0,-2, head_width = .4, head_length = .3, color = 'blue')
arrow(7,-1,0,2, head_width = .4, head_length = .3, color = 'blue')
arrow(7,-3.2,0,-2, head_width = .4, head_length = .3, color = 'blue')
xlabel('r')
ylabel('$x^*$')
title('Bifurcation diagram of $\dot{x}=r-x^2$')
```

# Example 2

This is a more complicated example:
$$
\mtab \dot{x}=r + \frac{x}{2} - \frac{x}{1+x}
$$
For an initial analysis, notice that the condition for an equilibrium can be written as:
$$
\mtab r + \frac{x}{2} = \frac{x}{1+x},
$$
which can be interpreted as intersecting the line $r + x/2$ and the curve $x/(1+x)$. The slope of the line is always $1/2$, so the parameter $r$ just determines how high the line is located.

A graphical analysis shows that we can have several cases:

```python
xmax = 9
xmin = 1-xmax 
x1 = linspace(-0.999,xmax,100)
y1 = x1/(1+x1)
x2 = linspace(xmin,-1.001,100)
y2 = x2/(1+x2)
axis([xmin,xmax,-8,8])
plot(x1,y1,x2,y2, linewidth = 2, color = 'blue')
x = linspace(xmin,xmax,200)
rvals = [-3,1.5,4.5]
for r in rvals:
    plot(x, r + x/2, linewidth = 2, color = 'red')
rvals = [0.07,2.9]
for r in rvals:
    plot(x, r + x/2, linewidth = 2, color = 'green')
```

The system can have 0, 1 or two equilibria, depending on the value of $r$. The bifurcations (there are two of them) happen when the line is tangent to the curve (the green lines in the picture). To find the values of $r$, we have to solve the tangency condition:
$$
\mtab\frac{d}{dx}\left[r+\frac{x}{2}\right] = \frac{d}{dx}\left[\frac{x}{1+x}\right]
$$
or
$$
\mtab\frac{1}{2} = \frac{1}{(1+x)^2}
$$
This gives two values (corresponding to the two bifurcations),
$ x=-1\pm\sqrt{2}$
To get the corresponding values of $r$, we plug in these values of $x$ in the equation for the equilibrium points, obtaining:

- $\displaystyle r^c_1 = \frac{3}{2} - \sqrt{2}$, for the equilibrium $x^*_1=-1+\sqrt{2}$. 
- $\displaystyle r^c_2 = \frac{3}{2} + \sqrt{2}$, for the equilibrium $x^*_2=-1-\sqrt{2}$.

Using these values, we can make a more accurate plot demonstrating the bifurcations:

```python
xmax = 9
xmin = 1-xmax 
x1 = linspace(-0.999,xmax,100)
y1 = x1/(1+x1)
x2 = linspace(xmin,-1.001,100)
y2 = x2/(1+x2)
axis([xmin,xmax,-8,8])
plot(x1,y1,x2,y2, linewidth = 2, color = 'blue')
x = linspace(xmin,xmax,200)
r1 = (3 - 2*sqrt(2))/2
r2 = (3 + 2*sqrt(2))/2
rvals = [r1, r2]
for r in rvals:
    plot(x, r + x/2, linewidth = 2, color = 'green')
delta = 2
rvals = [r1 - delta, (r1 + r2)/2, r2 + delta]
for r in rvals:
    plot(x, r + x/2, linewidth = 2, color = 'red')
xlabel('x')
title('Bifurcations of $\dot{x} = r - x/2 - x/(1+x)$')
```

What about stability? If the line $r-x/2$ is above the curve $x/(1+x)$, then the field is moving to the right, otherwise it is moving to the left. Using this, we conclude:

- if $r<r^c_1$, there are two equilibria, the rightmost is unstable and the leftmost is stable.
- if $r=r^c_1$, there is one half-stable equilibrium.
- if $r^c_1 < r < r^c_2$, there are no equilibria.
- if $r > r^c_2$, there are two equilibria, the rightmost is stable and the leftmost is unstable.

Let's now draw an accurate bifurcation diagram. We use the information above to determine which branches correspond to unstable/stable equilibria. Notice that there is a singularity in the field at $x=-1$ that makes things slightly complicated.

```python
r1, x1 = (3 - 2*sqrt(2))/2, -1 + sqrt(2)
r2, x2  = (3 + 2*sqrt(2))/2, -1 - sqrt(2)
xmax = 10
xmin = -10
rmin = -6
rmax = 8
axis([rmin, rmax, xmin, xmax])
x = linspace(x1, xmax, 100)
r = -x / 2 + x / (x + 1)
plot(r, x, '--', linewidth = 2, color = 'red')
x = linspace(-0.999, x1, 100)
r = -x / 2 + x / (x + 1)
plot(r, x, linewidth = 2, color = 'red')
x = linspace(xmin, x2, 100)
r = -x / 2 + x / (x + 1)
plot(r, x, '--', linewidth = 2, color = 'red')
x = linspace(x2, -1.001, 100)
r = -x / 2 + x / (x + 1)
plot(r, x, linewidth = 2, color = 'red')
# Decorations
xlabel('$r$')
ylabel('$x^*$')
title('Bifurcation diagram for $\dot{x}=r+x/2-x/(1+x)$')
axvline(x = r1, ymin = 0, ymax = (x1-xmin)/(xmax-xmin), linestyle = ':', color = 'black')
axvline(x = r2, ymin = 0, ymax = (x2-xmin)/(xmax-xmin), linestyle = ':', color = 'black')
axhline(y = x1, xmin = 0, xmax = (r1-rmin)/(rmax-rmin), linestyle = ':', color = 'black')
axhline(y = x2, xmin = 0, xmax = (r2-rmin)/(rmax-rmin), linestyle = ':', color = 'black')
axhline(y = -1, linestyle = '--', color = 'blue')
annotate('$r^c_1=3/2-\sqrt{2}$', xy = (r1,xmin), xytext = (r1 - 0.2, xmin + 1.5), 
          arrowprops = dict(arrowstyle='->'), horizontalalignment = 'right', verticalalignment='bottom')
annotate('$r^c_2=3/2+\sqrt{2}$', xy = (r2,xmin), xytext = (r2 + 0.2, xmin + 1.5), 
         arrowprops = dict(arrowstyle='->'), horizontalalignment = 'left', verticalalignment='bottom')
annotate('$x^*_1=-1+\sqrt{2}$', xy = (rmin, x1), xytext = (rmin + 1, x1 + 1), arrowprops = dict(arrowstyle='->'))
annotate('$x^*_2=-1-\sqrt{2}$', xy = (rmin, x2), xytext = (rmin + 1, x2 - 1), arrowprops = dict(arrowstyle='->'),
         verticalalignment = 'top')
```

The blue dashed line indicates the singularity of the field at $x=-1$.

Suppose now we want to display solution curves for a particular value of the parameter $r$. It is possible to solve the equation by separation of variables, but the solution is cumbersome (to say the least). So, we will produce numerical solutions.

The first step is to construct a function that defines the flow, that is a function that represents the map:
$$
\mtab (x,t) \to r + \frac{x}{2} - \frac{x}{1+x}
$$
Notice that we have to specify the independent variable $t$, even though the DE is autonomous.

We make a function factory to create the field for an arbitrary value of the parameter $r$:

```python
def make_de(r):
    de = lambda x, t : r + 0.5*x - x / (1 + x)
    return de
```

Now, let's import the function `odeint` from the package `scipy.integrate`:

```python
from scipy.integrate import odeint
```

In the next cell, we define the flow for the parameter $r=4$ and compute several solutions with different initial conditions. The syntax for the `odeint` function is:

    odeint(de, x0, t)
    
where:

- `de` is a Python function that takes two variables `x`, `t` as input, and returns the value of the field at `(x,t)`.
- `x0` is the initial condition.
- `t` is a vector of values at which the solution is computed.

```python
de = make_de(4)
t = arange(0,4.0,.01)
x1 = odeint(de, -3, t)
x2 = odeint(de,  -1.1, t)
x3 = odeint(de, -4,t)
x4 = odeint(de, -6, t)
x5 = odeint(de, -8, t)
```

Finally, we plot the solutions:

```python
axis([0, 4, -10, 0])
plot(t,x1, t, x2, t, x3, t, x4, t, x5, color = 'blue')
xlabel('$t$')
ylabel('$x(t)$')
title('Solution curves for the equation with $r=4$')
```

What if we want good approximations for the equilibrium points for a fixed value of the parameter? We have to solve the equation
$$
\mtab r +\frac{x}{2}-\frac{x}{x+1}=0.
$$
We first plot the function defining the flow, so that we get an idea of where the equilibrium points are:

```python
x = linspace(-7, -1.1, 100)
axis(ymin = -3, ymax = 3)
plot(x, de(x, 0), color = 'red')
axhline(0, color = 'black')
xlabel('x')
ylabel('\dot{x}')
title('Flow for $r=4$')
```

As expected, there are two equilibrium points, one between $-6$ and $-5$, the other between $-2$ and $-1$. We will use the function `brentq` from the package `scipy.optimize` to solve the equation. The syntax is the following:

    brentq(eq, a, b)

where:

- `eq` is the function defining the equation.
- `a` and `b` are an interval containing the root.

This is what we get:

```python
from scipy.optimize import brentq
eq = lambda x : de(x,0) 
xeq1 = brentq(eq, -6, -5)
xeq2 = brentq(eq, -2, -1.1)
print(xeq1, xeq2)
```

Let's now add the equilibrium solutions to the solution plot:

```python
axis([0, 4, -10, 0])
plot(t,x1, t, x2, t, x3, t, x4, t, x5, color = 'blue')
axhline(xeq1, color = 'red')
axhline(xeq2, color = 'red')
xlabel('$t$')
ylabel('$x(t)$')
title('Solution curves for the equation with $r=4$')
```
