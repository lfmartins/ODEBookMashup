```python
%pylab inline
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D
```

```python
def make_lorenz(sigma, r, b):
    def func(statevec, t):
        x, y, z = statevec
        return [ sigma * (y - x),
                 r * x - y - x * z,
                 x * y - b * z ]
    return func
```

```python
lorenz_eq = make_lorenz(10., 28., 8./3.)
```

```python
def ode_func(xvec, t, sigma, rho, beta):
    x, y, z = xvec
    return array([sigma * (y - x),
                  x * (rho - z) - y,
                  x * y - beta * z ])
```

```python
tmax = 50
tdelta = 0.005
tvalues = arange(0, tmax, tdelta) 
ic = array([0.0, 1.0, 0.0])
sol = odeint(ode_func, ic, tvalues, args=(10., 28., 8./3.))
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
x, y, z = array(zip(*sol))
ax.plot(x, y, z, lw=0.5, color='DarkBlue')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
None
```

```python
vmin = 1.
vmax = 2.
npoints = 1000
values = linspace(vmin, vmax, npoints)
xpoints = zeros(npoints)
ypoints = zeros(npoints)
zpoints = zeros(npoints)
tmax = 50.
tvalues = linspace(0,tmax,100)
for i, y0 in enumerate(values):
    ic = array([0.0, y0, 0.0], dtype=float64)
    sol = odeint(ode_func,  ic, tvalues, args=(10., 28., 8./3.))[-1]
    xpoints[i], ypoints[i], zpoints[i] = sol
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xpoints, ypoints, zpoints, color='red')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
```
