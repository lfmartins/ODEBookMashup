```python
from scipy.integrate import odeint
def solve_de(sysfunc, y0, t0, tmin, tmax, npoints=200, **kwargs):
    tleft = linspace(t0, tmin, npoints)
    yleft = odeint(sysfunc, y0, tleft, **kwargs)
    tright = linspace(t0, tmax, npoints)
    yright = odeint(sysfunc, y0, tright, **kwargs)
    return concatenate((yleft[-1:0:-1], yright))
```

```python
def make_linear_system(a,b,c,d):
    def sysfunc(xvec, _):
        x, y = xvec
        return array([a * x + b * y, c * x + d * y])
    return sysfunc
```

```python
# Options for quiver() function.
# See documentation of quiver() for meaning of parameters.
quiver_opts = dict(units='x',
                   scale=20,
                   width=0.03,
                   headwidth=3,
                   headlength=1.5,
                   headaxislength=1.5,
                   color='#3366FF',
                   alpha=1.0)
# These define the range of
# the plot, and the interval
# at which arrows are plotted.
xyrange = 2                                     
r = arange(-xyrange-1, xyrange+1, 0.2)
X, Y = meshgrid( r, r )
# System coefficients
a, b, c, d = -2, 1, 0, -1
# Initial conditions
ics = [[0,1],[0.5,1],[1,0.5],[1,-0.5],          
       [1,-1],[1,-2],[-1,-1.5],
       [-2,-1],[-2,1]]
# Time range
t0, tmin, tmax = 0, -10, 10
# Number of points in solution
npoints = 200
# Define linear system
sysfunc = make_linear_system(a, b, c, d)
# Define figure and axes
fig = figure(figsize=(8,8))
ax = fig.add_subplot(111, aspect='equal')
axis([-xyrange, xyrange, -xyrange, xyrange])
# Plot direction field
U, V = a * X + b * Y, c * X + d * Y
Q = quiver( X, Y, U, V, **quiver_opts)
# Plot eigendirections, if any.
A = array([[a, b],[c,d]], dtype=float64)
title_str = ''
eigvals, eigvecs = eig(A)
eigvecs = array(zip(*eigvecs))
for i, (eigval, eigvec) in enumerate(zip(eigvals, eigvecs)):
    title_str += '$\lambda_{}={},\ \mathbf{{v}}_{}=({:.5g},\ {:.5g})\quad$'.\
                 format(i+1, eigval, i+1, eigvec[0], eigvec[1])
    if abs(eigval.imag) < 1e-7:
        vec = (5*xyrange / norm(eigvec)) * eigvec
        plot( *zip(-vec, vec), lw=2, color='#009900')
# Plot solutions
for y0 in ics:
    y = solve_de(sysfunc, y0, t0, tmin, tmax, npoints=npoints)
    plot(*zip(*y), lw=2, color='#CC0000', aa=True)
xlabel('x')
ylabel('y')
title(title_str)
```

```python
tau=linspace(-4,4,200)
delta = tau ** 2 /4.0
arrow_props = dict(facecolor='purple',
                   shrink=0.04,
                   width=3, 
                   headwidth=6, 
                   frac=0.3)
fig=figure(figsize=(8,8))
ax = fig.add_subplot(111)
axis([-1,2,-4,4])
plot(delta,tau, color='g', lw=3, aa=True, label=r'$\tau^2-4\Delta=0$')
axvline(0,lw=2,color='b')
axhline(0,lw=2,color='b')
xlabel(r'$\Delta$')
ylabel(r'$\tau$')
text(-.75, 2, 'Saddles')
text(-.75, -2, 'Saddles')
annotate('Saddles', xy=(-0.35,0.03), xytext=(-0.75, 0.25),
          arrowprops=arrow_props)
text(1.0, -3, 'Stable nodes with')
text(1.0, -3.2, 'two eigendirections')
text(1.0, 3, 'Unstable nodes with')
text(1.0, 2.8,'two eigendirections')
text(0.35, 0.5,'Unstable spiral nodes')
text(0.35, -0.5,'Stable spiral nodes')
annotate('Eigendirection of', xy=(0,3.5), xytext=(0.2,3), 
         arrowprops=arrow_props)
text(0.2, 2.8, 'equilibrium points')
annotate('Eigendirection of', xy=(0,-3.5), xytext=(0.2,-3), 
         arrowprops=arrow_props)
text(0.2, -3.2, 'equilibrium points')
annotate('Centers', xy = (1.8,0), xytext=(1.35, 0.35),
         arrowprops=arrow_props)
annotate('Stable star/degenerate nodes', xy=(0.8,-1.75), xytext=(0.9, -1.5),
          arrowprops=arrow_props)
annotate('Unstable star/degenerate nodes', xy=(0.8,1.75), xytext=(0.9, 1.5),
          arrowprops=arrow_props)
ax.set_xticklabels([])
ax.set_xticks([])
ax.set_yticklabels([])
ax.set_yticks([])
legend()
title('The trace-determinant plane')
```
