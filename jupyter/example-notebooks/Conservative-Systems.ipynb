## Double-Well Potential
The system of differential equations is:
$$
\newcommand{\mtab}{\quad\quad\quad\quad}
\mtab \dot{x} = y
$$
$$
\mtab\dot{y} = x - x^3
$$
The conserved quantity is given by:
$$
\mtab E = \frac{1}{2}y^2 - \frac{1}{2}x^2+\frac{1}{4}x^4
$$
We first define a function representing the energy:

```python
def E(x, y):
    return 0.5 * y ** 2 - 0.5 * x ** 2 + 0.25 * x ** 4
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca(projection='3d')
X = arange(-2, 2, 0.1)
Y = arange(-1.5, 1.5, 0.1)
X, Y = meshgrid(X, Y)
Z = E(X, Y)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=True)
ax.set_zlim(0, 1)
plt.show()
```

Contours

```python
x = arange(-2.0, 2.0, 0.01)
y = arange(-2.5, 2.5, 0.01)
X, Y = np.meshgrid(x, y)
Z = E(X, Y)
figure(figsize=(8,8))
CS = plt.contour(X, Y, Z, [ -.5, -0.1, 0, 1.0, 1.5, 2.0], colors = 'blue', lw = 2)
#plt.clabel(CS, inline=1, fontsize=10)
```

# Predator-prey

The conserved quatity is:
$$
\mtab E =  dR + cF - a\ln(F)-c\ln(R)
$$
The contour plots are:

```python
a = 1
b = 0.2
c = 2
d = 0.5
x = arange(0.1, 10, 0.01)
y = arange(0.01, 1.8, 0.01)
X, Y = np.meshgrid(x, y)
Z = d * X + c * Y - a * log(Y) - c * log(X)
figure(figsize=(8,8))
CS = plt.contour(X, Y, Z, [1,1.2, 1.4, 1.6, 1.8, 2], colors = 'blue', lw = 2)
#plt.clabel(CS, inline=1, fontsize=10)
```

## Pendulum (undamped)

The enerigy function is:
$$
\mtab E = \frac{1}{2}v^2-\cos(\theta)
$$

```python
x = arange(-3*pi, 3*pi, 0.01)
y = arange(-3*pi, 3*pi, 0.01)
X, Y = np.meshgrid(x, y)
Z = 0.5 * Y ** 2 - cos(X)
figure(figsize=(8,8))
CS = plt.contour(X, Y, Z, [ -0.5, 0.25, 1, 2, 4, 6], colors = 'blue', lw = 2)
#plt.clabel(CS, inline=1, fontsize=10)
```
