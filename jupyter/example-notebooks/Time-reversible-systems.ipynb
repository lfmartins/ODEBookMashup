## Example 1

$$
\newcommand{\mtab}{\quad\quad\quad\quad}
\mtab\dot{x}=y-y^3
$$
$$
\mtab\dot{y}=-x-y^2
$$
We start by defining a function representing the system:

```python
def sysfunc(xvec, t):
    x, y = xvec
    return y - y ** 3, -x - y ** 2
```

Before we plot solution curves, let's do some analysis. The equilibrium points are $(0,0)$, $(-1,-1)$ and $(-1,1)$. The Jacobian matrix of the system is:
$$
\mtab
\begin{pmatrix}
0  & 1-3y^2\\
-1 & -2y\\
\end{pmatrix}
$$
Just for convenience, we can also define a function that computes the jacobian:

```python
def jacobian(x, y):
    return matrix([[0, 1 - 3 * y ** 2],
                   [-1, -2 * y]], dtype=float64)
```

Let's now do the analysis at each equilibrium point:
$$
(0,0):
$$

```python
J = jacobian(0,0)
print(J)
eig(J)
```

So, the linearized system is a _center_. Since the nonlinear system is time-reversible, this implies that the origin is a center in the nonlinear system too.

```python
J = jacobian(-1,1)
print J
eig(J)
```

The linearized system is a _saddle_. The stable and unstable manifolds are tangent to the vectors :
$$
\mtab
\begin{pmatrix}
0.59 \\ 0.81
\end{pmatrix}
$$
and
$$
\mtab
\begin{pmatrix}
0.94 \\ -0.34
\end{pmatrix}
$$
(approximating to 2 decimals, of course).

Since the system is time-reversible, by symmetry we know that $(-1,-1)$ will also be a saddle. The stable manifold of $(-1,-1)$ is the reflection of the unstable manifold of $(-1,1)$ and vice-versa. Let's now plot the phase plane. We first define the following auxiliary function to compute solutions:

```python
from scipy.integrate import odeint
def solve_de(sysfunc, y0, t0, tmin, tmax, npoints=200, **kwargs):
    tleft = linspace(t0, tmin, npoints)
    yleft = odeint(sysfunc, y0, tleft, **kwargs)
    tright = linspace(t0, tmax, npoints)
    yright = odeint(sysfunc, y0, tright, **kwargs)
    return concatenate((yleft[-1:0:-1], yright))
```

```python
# Options for quiver() function.
# See documentation of quiver() for meaning of parameters.
quiver_opts = dict(units='x',
                   scale=20,
                   width=0.03,
                   headwidth=3,
                   headlength=1.5,
                   headaxislength=1.5,
                   color='#3366FF',
                   alpha=1.0)
# These define the range of
# the plot, and the interval
# at which arrows are plotted.
xyrange = 2                                    
r = arange(-xyrange-1, xyrange+1, 0.2)
X, Y = meshgrid( r, r )
# Initial conditions
# Specify time range for each solution separately
delta = 1E-2
ics = [[[-1+delta*0.9390708, 1-delta*0.34372377], -0.1, 8.7],
       [[-1+delta*0.9390708, -1+delta*0.34372377], -8.7, 0.1],
       [[-1-delta*0.9390708, 1+delta*0.34372377], -0.1, 7],
       [[-1-delta*0.9390708, -1-delta*0.34372377], -7, 0.1],
       [[-1+delta*0.59069049, 1-delta*0.80689822], -5, 0.1],
       [[-1-delta*0.59069049, 1+delta*0.80689822], -2, 0.1],
       [[-1-delta*0.59069049, -1-delta*0.80689822], 0.1, 2],
       [[-1,0], -0.1, 9],
       [[-0.5,0], -0.1, 9],
       ]
# Number of points in solution
npoints = 200
# Define figure and axes
fig = figure(figsize=(8,8))
ax = fig.add_subplot(111)
axis([-xyrange, xyrange, -xyrange, xyrange])
# Plot direction field
U, V = sysfunc([X, Y], 0)
Q = quiver( X, Y, U, V, **quiver_opts)
for y0, tmin, tmax in ics:
    y = solve_de(sysfunc, y0, 0, tmin, tmax, npoints=npoints)
    plot(*zip(*y), lw=2, color='#CC0000', aa=True)
    pass
xlabel('x')
ylabel('y')
```

Notice how we plotted an approximation to the stable/unstable manifolds. Feel free to experiment with other initial conditions.
