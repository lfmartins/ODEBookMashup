```python
from scipy.integrate import odeint
def solve_de(sysfunc, y0, t0, tmin, tmax, npoints=200, **kwargs):
    tleft = linspace(t0, tmin, npoints)
    yleft = odeint(sysfunc, y0, tleft, **kwargs)
    tright = linspace(t0, tmax, npoints)
    yright = odeint(sysfunc, y0, tright, **kwargs)
    return concatenate((yleft[-1:0:-1], yright))
```

```python
def make_sysfunc(alpha,I):
    def f(xvec,t):
        x, y = xvec
        return array([ y,
                       I - sin(x) - alpha * y])
    return f
```

```python
sysfunc = make_sysfunc(0.5, 0.75)
```

```python
alpha = 0.5
I = 0.75
sysfunc = make_sysfunc(alpha, I)
xmax = 2 * pi
ymax = 3.
# Initial conditions
# Specify time range for each solution
ics = [[[0, 1], -5, 15],
       [[2*pi, 1], -5, 1.1],
       [[2.28, 0], -10, 15],
       [[2.31, 0], -10, 10],
       [[0, 2.5], -0.1, 10],
       ]
# Number of points in solution
npoints = 200
# Define figure and axes
fig = figure(figsize=(8,8))
ax = fig.add_subplot(111)
axis([0.0, xmax, -ymax, ymax])
t0 = 0
for y0, tmin, tmax in ics:
    y = solve_de(sysfunc, y0, t0, tmin, tmax, npoints=npoints)
    plot(*zip(*y), lw=2, color='#CC0000', aa=True)
    pass
xlabel(r'$\phi$')
ylabel(r'$y$')
title(r'Forced pendulum')
```

```python
alpha = 0.5
I = 1.5
sysfunc = make_sysfunc(alpha, I)
```

```python
tmax = 30
deltat = 0.1
tvalues = arange(0, tmax + deltat, deltat)
def poincare(y0):
    svalues = odeint(sysfunc, array([0.,y0]), tvalues)
    idx = nonzero(svalues[:,0] > 2*pi)[0][0]
    return svalues[idx,1]
```

```python
poincare(3.0)
```

```python
yvalues = arange(0,5,0.1)
Pvalues = array([poincare(y0) for y0 in yvalues])
```
